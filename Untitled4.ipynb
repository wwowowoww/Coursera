```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

# 日付の指定
yyyy = input("調査する年（西暦４桁）を入力してください。: ")
mmdd = input("調査する月日（mmdd）を入力してください。: ")

# 時刻の指定
# 時刻は0,3,6,9,12,15,18,21のいずれかを選択する。
HH = int(input("調査する時刻（0, 3, 6, 9, 12, 15, 18, 21）を入力してください。: "))

url = f"http://database.rish.kyoto-u.ac.jp/arch/jmadata/data/gpv/netcdf/MSM-P/{yyyy}/{mmdd}.nc"

# curlコマンドを実行してファイルをダウンロード
!curl -O $url

ds = xr.open_dataset(f"{mmdd}.nc")

plt.rcParams["font.size"] = 18

# 相模湾の緯度経度を指定して拡大表示
lon_min, lon_max, lat_min, lat_max = 138.5, 140, 34.5, 35.5
skip = 1
plev = 850
qx, qy, qk = 1.2, 0.05, 5

ax = plt.axes(projection=ccrs.PlateCarree())

# mmddから月(mm)と日(dd)を分離
mm = mmdd[:2]
dd = mmdd[2:]

time_string = f"{yyyy}-{mm}-{dd} {HH:02}:00:00"
u = ds.u.sel(p=plev).sel(time=time_string)[::skip, ::skip]
v = ds.v.sel(p=plev).sel(time=time_string)[::skip, ::skip]

#scaleの数字で矢印の長さを調整可能
q = ax.quiver(ds.lon[::skip], ds.lat[::skip], u, v, scale=150, transform=ccrs.PlateCarree())
ax.quiverkey(q, X=0.9, Y=1.05, U=qk, label=f"{qk} m/s", labelpos='E') # X, Y の値を調整

ax.set_title(ds.sel(time=time_string).time.dt.strftime("%Y-%m-%d %HUTC").values)
ax.coastlines()
ax.gridlines()
ax.set_extent([lon_min, lon_max, lat_min, lat_max])  # 地図の拡大表示

plt.show()
```
